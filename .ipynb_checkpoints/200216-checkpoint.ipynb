```python
# ※このCSVファイルは、「振動DB研究会計測船まとめ(20190201)PWD.xls」をCSVに変換し、2次偶力の単位がton・mのデータに印をつけたもの。
data = '../0_data/measuredData.csv'
var = ['no.', 'shipyard', 'year/month', 'type', 'classification', 'Lpp', 'B', 'D', 'd', 'D/W', 'Hold', 'engineType', 'cylinder', 'NOR', 'MCR', 'secondaryCouple', 'exc', 'balancer', 'frequency2', 'frequency3', 'frequency4', 'frequency5', 'frequency6', 'frequencySuperstructure' ,'frequencyChimney', 'chimneyHeight', 'chimneyPresence', 'exc', 'exc', 'exc']
df = pd.read_csv(data, names=var, skiprows=10, index_col=0, skipfooter=1, engine='python')
df = df.reset_index(drop=True)

# 表示
pd.set_option('display.max_columns', 50)
pd.set_option('display.max_rows', 150)

# 単位の統一
# tom・m単位のところは右側に'convert'と標記
for x in range(len(df)):
    if df.loc[x, 'exc'] == 'convert':
        df.loc[x, 'secondaryCouple'] = df.loc[x, 'secondaryCouple'] * 9.8

# 不要な列を削除
var_unnecessary = ['shipyard', 'year/month', 'classification', 'exc', 'chimneyPresence', 'exc.1', 'exc.2', 'exc.3']
# df_organized = df.drop(columns=var_unnecessary)
df = df.drop(var_unnecessary, axis=1)

# 後々追加するかもしれないが、今回は削除する変数
var_unnecessary2 = ['engineType']
# df_organized = df.drop(columns=var_unnecessary2)
df = df.drop(var_unnecessary2, axis=1)

# イレギュラーな入力を整える
df = df.replace(['-', 'ー', '－', '？', '?'], np.nan).replace({'858?': 858, '5（7）': 5, '10(P/S)': 10, '蒸気': np.nan})

# カテゴリ変素をダミー化
df_organized = pd.get_dummies(df, columns=['type', 'balancer'])

# 文字列型の数字データをfloat型に変換
for column_name in df_organized:
    df_organized[column_name] = df_organized[column_name].astype(float)

df_organized_onlyFrequency2 = df_organized.drop(['frequency3', 'frequency4', 'frequency5', 'frequency6', 'frequencySuperstructure', 'frequencyChimney'], axis=1)
    
df_organized_onlyFrequency2

```

### 決定係数

#### 決定係数（$R^2$、アールツースコア）  

**特徴**

- モデルによって予測した値が実際の値とどの程度一致しているかを表現する評価指標。
- 説明変数が目的変数をどのくらい説明できるかを表す。
- 異なる回帰モデル間での良し悪しを判断するのに使う。


**評価方法**
- 一概には言えないが、0.7~0.8くらいほしい
- 0.9を超えてしまうと過学習の可能性も考えられる？
- この値が高ければ絶対に良いわけではないので、他の指標も含めて総合的に評価する必要がある
- testデータとtrainデータ両方の決定係数を出し、比較する。→testデータの決定係数がtrainデータのそれよりも大きく下回るようであれば、過学習を起こしている。（差を0.05程度？に抑えたい。）

$$
R^2 = 1 -  \frac{\displaystyle \sum_{ i   }^{  }( 予測値_i - 実測値_i )^2 }{\displaystyle \sum_{ i   }^{  }( 予測値_i - 実測値の平均値 )^2}
$$

```python
score_rfr_train = rfr.score(X_train, y_train)
score_rfr_test = rfr.score(X_test, y_test)
score_svr_train = svr.score(X_train, y_train)
score_svr_test = svr.score(X_test, y_test)

print('Random Forest Regression')
print('決定係数（train）: {}'.format(score_rfr_train))
print('決定係数（test）: {}'.format(score_rfr_test))
print('-----------------------------------')
print('Support Vector Regression')
print('決定係数（train）: {}'.format(score_svr_train))
print('決定係数（test）: {}'.format(score_svr_test))
```

#### 自由度調整済み決定係数（$R_f^2$）  

**特徴**

- 決定係数は説明変数の数が増えるほど1に近づくという性質を持っている。→これを改善したのが自由度調整済み決定係数。
- 説明変数が多いときや、説明変数の数が異なる回帰モデルどうしを比較する場合に有利。

**評価方法**

- 決定係数と同じ。

$$
R^2 = 1 -  \frac{\frac{\displaystyle \sum_{ i   }^{  }( 予測値_i - 実測値_i )^2}{n - k - 1} }{\frac{\displaystyle \sum_{ i   }^{  }( 予測値_i - 実測値の平均値 )^2}{n - 1}}
\\ （n: サンプルサイズ、k: 説明変数の数）
$$
